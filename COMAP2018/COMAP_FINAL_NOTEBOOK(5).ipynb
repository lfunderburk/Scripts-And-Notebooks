### Abstract 

Energy production and consumption form important components of a given economy. On a national level, aspects of energy policy can be either centralized or descentralized. Aspects that make policy creation and implementation challenging are the geographical and industrial differences that exist within the same nation, as these affect energy production and consumption on both a local and national level. 

One approach to this problem is forming an interstate compact, that is, a contractual agreement made between two or more entities whose goal is to agree on a specific policy issue and either adopt a set of standards or cooperate with one another. 

### Problem 

In this notebook I will explore a particular instance of this problem. Consider the United States and the four states California, Arizona, New Mexico and Texas. They come together and wish to form a realistic new energy compact whose focus is to increase the use of cleaner, renewable energy sources. I am given 50 years (from 1960 to 2009) worth of data in 605 variables on each of the states' production and consumption, in addition to demographic and economic information.

### Strategy 

This problem can be seen from a game theoretic point of view, where each state is a player whose actions are either: cooperate, remain stagnant, apply internal policy change. 

Step 1) With the data, create a profile for each state based on consumption and production on two levels: on a given year (2009 for example) and throughout the time frame 1960 - 2009.

Step 2) Model a score scheme that will allow to determine who the dominant and weak players are and apply scoring scheme on the levels described above. 

Step 3) Use score to determine who dominant and weak players are over time.

Step 4) With the help of our model, predict in 2025 and 2050 where each state will go without policy change. 

Step 5) Propose a policy change that prompts weak players to apply internal changes in their state, and dominant players to help weak players while they transition from fossil fuels to renewable energy sources.

### Scoring Scheme 

The determination of best and wost preforming states was done by looking at the energy consumption in each state and examining the carbon output of larger fossil fuel categories. The specific categories of fossil fuel consumption are: Coal, Distillate, Motor Fuels, Jet Fuels, LPG, and Natural Gas. For this model we only considered consumption of fossil fuels. Production of energy was not considered in this scoring to not punish a state for producing the "dirty" energy used by another and too not overly punish a state for what has become an, unfortunate and immediate, economic necessity.

We model the scoring scheme as follows.  

Let $S$ denote score for a given state, let $x_i$ (units of $CO_2$ per billion Btu) denote the amount of carbon diaoxide emissions produced from consuming fossil fuels, where $i \in \lbrace C,N,M,D,J,P,R \rbrace$ and each letter denotes, respectively, Coal, Natural Gas, Motor Gasoline, Distillate, Jet Fuet, Propane/LPG, Residual. Similarly, let $y_i$,  $i \in \lbrace C,N,M,D,J,P,R \rbrace$, denote the amount of fossil fuels consumed in that state. Finally, let $P$ the total population$^{[1]}$. For reading purposes, we divide the score by 1000 and round to the nearest tenth, although we did not perform such computation on the data. We then compute the score as follows:


$$S = \dfrac{1}{P} \sum _{i \in \lbrace C,N,M,D,J,P,R \rbrace}x_i \times y_i$$


The values of $C$ were based on the table $^{[2]}$ below:

The values of $D$ were obtained from the data provided by COMAP and based on our selection of variables, while the populations were obtained from $[1]$.

### Proposed Lines of Action

We propose the following three lines of action.

#### Carbon tax

A Carbon tax is a system of taxation placed on the purchase of fuel or energy that has carbon emission attached to it. By increasing the cost of carbon intensive energy forms, the tax attempts to influence purchasing decisions in favour of "greener" forms of energy. The effectiveness of this tax to influence purchasing decisions is influenced heavily by the availability of "greener" alternatives, but the tax could help to stimulate the development of such a market as well. 

For the objective of improving the score defined in this paper the Tax would be on kg of carbon produced per BTU of fuel burned as defined the the U.S. Energy Information Agency.

#### Strong Helps Weak 

Implement a policy which allows strong states to support weak ones during the transition period. Policies may include exchange of credits and/or goods. 

#### Cap and Trade 

This is a more complicated system then a straight carbon tax, but in other ways more direct. Cap and Trade sets a cap on carbon emissions for industry or for the economy. The government then distributes credits to companies that allow them to emit $CO_2$ up to the limit of their credit. To go above that a company would need to buy carbon credits on the open market from another company that has not used all theirs, risk paying a hefty fine for being in breach of the law, or cease operations in order to comply with the law and not emit any more $CO_2$.
